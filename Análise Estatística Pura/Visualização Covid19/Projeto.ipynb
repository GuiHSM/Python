```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range (len(y)):
            datasets.append({
                'label': labels[i],
                'data':y[i]
            })
            print(datasets)
        return datasets
    return [
        {
            'label': labels[0],
            'data':y
        }
    ]
```

```python
def set_title(title=''):
    if title != "":
        display="true"
    else:
        display="false"
    return {
        'title': title,
        'display' : display
    }
```

```python
def create_chart(x,y,labels,kind="bar",title=''):
    datasets = get_datasets(y,labels)
    options = set_title(title)
    chart = {
        'type':kind,
        "data":{
            'labels':x,
            'datasets':datasets
        },
        'options':options
    }
    
    return chart
```

```python
def get_api_chart(chart):
    url_base = "https://quickchart.io/chart"
    requisition=r.get(f'{url_base}?c={str(chart)}')
    return requisition.content
```

```python
def save_image(path,content):
    with open(path,'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pil= Image.open(path)
    display(img_pil)
```

```python
#Os dados de Recuperação vão só até dia 30/07/2021 mais ou menos, então por isso o gráfico só vai até esse dia
y_data_1=[]
y_data_2=[]
x=[]
comecou=False
for obs in final_data[1::20]:
    if obs[Recuperados]==0 and comecou:
        break
    if obs[Recuperados]>0:
        comecou=True
    y_data_1.append(obs[Confirmados])
    y_data_2.append(obs[Recuperados])
    x.append(obs[Data].strftime("%d/%m/%Y"))
    
labels=["Confirmados","Recuperados"]

chart = create_chart(x, [y_data_1,y_data_2],labels,title="Gráfico ConvirmadoXRecuperados")
chart_content=get_api_chart(chart)
save_image("Grafico.png",chart_content)
display_image("Grafico.png")
```
